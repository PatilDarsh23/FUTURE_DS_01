```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("default")
```

```python
data = {
    "Order_ID": [101,102,103,104,105,106,107,108],
    "Product": ["Laptop","Phone","Tablet","Laptop","Phone","Tablet","Laptop","Phone"],
    "Category": ["Electronics"]*8,
    "Region": ["North","South","East","West","North","South","East","West"],
    "Sales": [50000,30000,20000,55000,32000,21000,60000,35000],
    "Month": ["Jan","Jan","Feb","Feb","Mar","Mar","Apr","Apr"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
total_sales = df["Sales"].sum()
print("Total Sales:", total_sales)
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
product_sales
```

```python
product_sales.plot(kind="bar", title="Sales by Product")
plt.ylabel("Sales Amount")
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum()
region_sales
```

```python
region_sales.plot(kind="pie", autopct="%1.1f%%", title="Region-wise Sales")
plt.ylabel("")
plt.show()
```

```python
monthly_sales = df.groupby("Month")["Sales"].sum()
monthly_sales
```

```python
monthly_sales.plot(kind="line", marker="o", title="Monthly Sales Trend")
plt.ylabel("Sales Amount")
plt.show()
```

## Key Insights

- Laptops generated the highest revenue among products.
- North and East regions performed better.
- Sales increased steadily from January to April.
- Electronics category dominates total revenue.
